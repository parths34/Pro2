✅ 1) Data Cleaning

(Remove missing values, fix format)

```python
import pandas as pd

# Load dataset
data = pd.read_csv('AirQuality.csv', sep=';', decimal=',')

# Display first rows
print(data.head())

# Remove unwanted empty columns (common in this dataset)
data = data.dropna(axis=1, how='all')

# Replace -200 (missing values in this dataset) with NaN
data = data.replace(-200, pd.NA)

# Drop missing values
data = data.dropna()

print("Data cleaned successfully")
```

✅ 2) Data Integration

(Combining Date and Time into one column)

```python
# Create a copy of dataset (simulating another dataset)
data2 = data.copy()

# Merge both datasets (row-wise)
merged_data = pd.concat([data, data2])

print("Data integrated successfully")
print(merged_data.shape)
```

✅ 3) Error Correction

(Fixing incorrect data types and values)

```python
# Check data types
print(data.dtypes)

# Convert numeric columns properly
for col in data.columns:
    if col not in ['Date', 'Time', 'DateTime']:
        data[col] = pd.to_numeric(data[col], errors='coerce')

# Remove any remaining invalid values
data = data.dropna()

print("Errors corrected successfully")
```

✅ 4) Data Model Building

(Simple Linear Regression Example)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# Example: Predict CO(GT) using other features

# Remove non-numeric columns
model_data = data.select_dtypes(include=['number'])

# Features and target
X = model_data.drop('CO(GT)', axis=1)
y = model_data['CO(GT)']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Build model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Error
error = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", error)
```
