```python
import pandas as pd

# Load dataset
df = pd.read_csv('Amazonbookreviews.csv')

# View first few rows
df.head()
```

```python
# =========================================================
# 1) CREATE DATA SUBSETS (ROW-WISE)
# =========================================================

# Example 1: First 100 rows
subset1 = df.iloc[:100]

# Example 2: Rows where rating >= 4 (assuming 'Score' column exists)
if 'Score' in df.columns:
    subset2 = df[df['Score'] >= 4]
else:
    subset2 = df.iloc[100:200]  # fallback

print("\nSubset 1 (First 100 rows):")
print(subset1.head())

print("\nSubset 2:")
print(subset2.head())
```

```python
# =========================================================
# 2) MERGE DATA (ROW-WISE)
# =========================================================

# Combine subset1 and subset2 row-wise
merged_data = pd.concat([subset1, subset2], axis=0)

print("\nMerged Data (Row-wise):")
print(merged_data.head())
```

```python
# 3) SORT DATA
# =========================================================

# Sort by Score (if exists), otherwise first column
if 'Score' in df.columns:
    sorted_data = df.sort_values(by='Score', ascending=False)
else:
    sorted_data = df.sort_values(by=df.columns[0])

print("\nSorted Data:")
print(sorted_data.head())
```

```python
# =========================================================
# 4) TRANSPOSE DATA
# =========================================================

# Transpose dataset (rows <-> columns)
transposed_data = df.T

print("\nTransposed Data:")
print(transposed_data.head())
```

```python
# =========================================================
# 5) RESHAPE DATA
# =========================================================

# Example 1: Melt (wide → long format)
if len(df.columns) >= 3:
    reshape_melt = pd.melt(df, id_vars=[df.columns[0]])
    print("\nReshaped Data using Melt:")
    print(reshape_melt.head())

# Example 2: Pivot (if suitable columns exist)
if 'Score' in df.columns and 'ProductId' in df.columns:
    reshape_pivot = df.pivot_table(values='Score', index='ProductId', aggfunc='mean')
    print("\nReshaped Data using Pivot:")
    print(reshape_pivot.head())
```

```python
# 6) CREATE DATA SUBSETS (COLUMN-WISE)
# =====================================================

subset_columns = df[["reviewerID", "asin", "overallRating"]]

print("\nColumn-wise Subset:")
print(subset_columns.head())
```

```python
# 6) MERGE DATA (COLUMN-WISE)
# =====================================================

col_subset1 = df[["reviewerID", "asin"]]
col_subset2 = df[["overallRating", "Summary"]]

merged_columns = pd.concat([col_subset1, col_subset2], axis=1)

print("\nMerged Column-wise Data:")
print(merged_columns.head())
```
