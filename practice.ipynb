```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Correct loading (IMPORTANT)
data = pd.read_csv('AirQuality.csv', sep=';', decimal=',')

# Remove empty columns
data = data.dropna(axis=1, how='all')
```

```python
plt.figure()
sns.boxplot(data=data['CO(GT)'])  # CO levels
plt.title("Box Plot of CO(GT)")
plt.show()
```

```python
plt.figure()
plt.hist(data['NO2(GT)'], bins=20)
plt.title("Histogram of NO2(GT)")
plt.xlabel("NO2 Levels")
plt.ylabel("Frequency")
plt.show()
```

```python

plt.figure()
plt.plot(data['CO(GT)'])
plt.title("CO(GT) Trend")
plt.xlabel("Index")
plt.ylabel("CO Levels")
plt.show()
```
