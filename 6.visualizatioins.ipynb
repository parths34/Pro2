```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('forestfires (1).csv')
```

```python
# Convert categorical months to ordered
month_order = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True)
```

```python
plt.figure()
sns.histplot(df['area'], bins=30, kde=True)
plt.title('Distribution of Burned Area')
plt.show()
```

```python
# 🔹 2. Fires per month
plt.figure()
sns.countplot(x='month', data=df)
plt.title('Number of Fires per Month')
plt.xticks(rotation=45)
plt.show()
```

```python
# 🔹 3. Temperature vs Area
plt.figure()
sns.scatterplot(x='temp', y='area', data=df)
plt.title('Temperature vs Burned Area')
plt.show()
```

```python
# 🔹 4. Correlation heatmap
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Load dataset
df = pd.read_csv('AirQuality.csv', sep=';', decimal=',')

# Handle missing values (replace -200 with NaN)
df.replace(-200, pd.NA, inplace=True)
```

```python
# 🔹 1. CO concentration distribution
# df.columns = df.columns.str.strip()
plt.figure()
sns.histplot(df['CO(GT)'].dropna(), kde=True)
plt.title('CO Concentration Distribution')
plt.show()
```

```python
# 🔹 2. Temperature vs Humidity
plt.figure()
sns.scatterplot(x='T', y='RH', data=df)
plt.title('Temperature vs Humidity')
plt.show()
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.columns = df.columns.str.strip()
df.columns = df.columns.str.strip()
df.replace(-200, pd.NA, inplace=True)
df['CO(GT)'] = pd.to_numeric(df['CO(GT)'], errors='coerce')
```

```python
plt.figure()
df['CO(GT)'].plot()
plt.title('CO Levels Over Time')
plt.xlabel('Index')
plt.ylabel('CO(GT)')
plt.show()
```

```python
# 🔹 4. Correlation heatmap
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df = pd.read_csv('heart.csv')
df.columns = df.columns.str.strip()

print(df.columns)  # verify

sns.countplot(x=df.columns[-1], data=df)  # last column is usually target
plt.title('Heart Disease Distribution')
plt.show()
```

```python
# 🔹 2. Age distribution
plt.figure()
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()

```

```python
# Clean columns
df.columns = df.columns.str.strip().str.lower()

# Detect columns
target_col = [col for col in df.columns if 'target' in col or 'output' in col][0]
chol_col = [col for col in df.columns if 'chol' in col][0]

# Plot
plt.figure()
sns.boxplot(x=target_col, y=chol_col, data=df)
plt.title('Cholesterol vs Heart Disease')
plt.show()
```
