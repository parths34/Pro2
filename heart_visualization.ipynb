✅ Step 1: Import & Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('heart.csv')

# View data
print(data.head())
```

📊 1. Box Plot (Check Outliers)

```python
plt.figure()
sns.boxplot(data=data['age'])   # you can also use 'chol', 'thalach'
plt.title("Box Plot of Age")
plt.show()
```

📊 2. Histogram (Distribution)

```python
plt.figure()
plt.hist(data['chol'], bins=20)
plt.title("Histogram of Cholesterol")
plt.xlabel("Cholesterol")
plt.ylabel("Frequency")
plt.show()
```

📈 3. Single Line Graph (Trend)

```python
plt.figure()
plt.plot(data['thalachh'])   # correct column name
plt.title("Heart Rate Trend")
plt.xlabel("Index")
plt.ylabel("Thalachh")
plt.show()
```

📈 4. Multiple Line Graph (Comparison)

```python
plt.figure()
plt.plot(data['age'], label='Age')
plt.plot(data['chol'], label='Cholesterol')
plt.plot(data['thalachh'], label='Thalachh')  # corrected

plt.title("Heart Data Comparison")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.show()
```
