```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('dataset_Facebook.csv', sep=';')

# Display first few rows
print(data.head())
```

📊 1. Box Plot (to detect outliers)

```python
plt.figure()
sns.boxplot(data=data['like'])  # 'like' column
plt.title("Box Plot of Likes")
plt.show()
```

📊 2. Histogram (to see frequency distribution)

```python
plt.figure()
plt.hist(data['like'], bins=20)
plt.title("Histogram of Likes")
plt.xlabel("Likes")
plt.ylabel("Frequency")
plt.show()
```

📈 3. Single Line Graph

```python
plt.figure()
plt.plot(data['like'])
plt.title("Single Line Graph of Likes")
plt.xlabel("Index")
plt.ylabel("Likes")
plt.show()
```

```python
📈 4. Multiple Line Graph
```

```python
plt.figure()
plt.plot(data['like'], label='Likes')
plt.plot(data['share'], label='Shares')
plt.plot(data['comment'], label='Comments')

plt.title("Multiple Line Graph")
plt.xlabel("Index")
plt.ylabel("Count")
plt.legend()
plt.show()
```
