✅ 1) Data Cleaning

(Removing missing values, duplicates)

```python
import pandas as pd

# Load dataset
data = pd.read_csv('heart.csv')

# Display first rows
print(data.head())

# Check missing values
print(data.isnull().sum())

# Remove missing values (if any)
data = data.dropna()

# Remove duplicate rows
data = data.drop_duplicates()

print("Data cleaned successfully")
```

✅ 2) Data Integration

(Here we simulate combining data — since usually only one dataset is given)

```python
# Create a copy of dataset (simulating another dataset)
data2 = data.copy()

# Merge both datasets
merged_data = pd.concat([data, data2])

print("Data integrated successfully")
print(merged_data.shape)
```

✅ 3) Error Correction

(Fixing wrong or inconsistent values)

```python
# Check basic statistics
print(data.describe())

# Example corrections:

# Age should not be negative
data = data[data['age'] > 0]

# Cholesterol should not be negative
data = data[data['chol'] > 0]

# Fix categorical values if needed
# Example: convert sex column to 0/1 if required
data['sex'] = data['sex'].replace({0: 0, 1: 1})

print("Errors corrected successfully")
```

✅ 4) Data Model Building

(Simple Machine Learning Model – Logistic Regression)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Features and target
X = data.drop('output', axis=1)
y = data['output']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Build model
model = LogisticRegression(max_iter=1000)

# Train model
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```
